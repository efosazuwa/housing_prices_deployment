**We can see that there are some columns that look they are missing a lot of their values. Let's look more closely.**

```python
nul_cols = df.apply("isnull").sum(axis=0)
for i,col in enumerate(nul_cols):
    if col:
        col = col/df.shape[0]*100
        print("{:13s} is missing {:.2f}% of its values".format(df.columns[i], col))
```

```python
df.loc[:,df.apply("isnull").sum(axis=0)!=0]
```

```python
def analyse_nulls(data, var):
    """
    Checks how mean Sales price Changes with a particular variable
    whether it is null or not.
    
    inputs:
        data: A dataframe
        var: A variable name from the same dataframe
    outputs:
        a Bar plot
    """
    data = data.copy()
    
    data[var] = np.where(data[var].isnull(), 1, 0)
    
    data.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    
    plt.show()
    
null_col_df = df.loc[:,df.apply("isnull").sum(axis=0)!=0]

for col in null_col_df:
    analyse_nulls(df, col)
```

There appear to be significant differences in sales price on whether a variable is missing or not for most of the variables with missing values. Which suggests there is potentially some important information missing.

### Data Exploration Numerical Values

```python
num_vars = [var for var in df.columns if df[var].dtypes != 'object']
```

```python
df[num_vars]
```

Some numerical values aren't suitable as predictive values such as 'ID' and others are categorical variables such as 'YrSold', 'MoSold', and 'OverallQual'.

```python
import re
for col in df.columns:
    if 'Yr', "Year" in col:
        print(col)
```

```python
YrVar = [col for col in df.columns if re.findall("Yr|Year", col)]
YrVar
```

```python
df.groupby("YrSold").median()['SalePrice'].plot()
```

We can see that the median price of houses declined from 2006 to 2010 which is unusual since property is the type of investment that appreciates over time. Could this be because the house is older? Could a few years difference result in a median house price $8000 cheaper?

```python
def age_v_price(data, var):
    """
    Compares the age of certain milestones concerning the house and it's sale price.
    
    inputs:
        data: A dataframe
        var: A variable name from the same dataframe
    outputs:
        a scatter plot
    """
    data = data.copy()
    
    data[var] = data['YrSold'] - data[var]
    
    plt.scatter(data[var], data['SalePrice'])
    plt.xlabel("Years since {}".format(var))
    plt.ylabel("SalePrice")
    plt.show()
    
for var in YrVar:
    if var != 'YrSold':
        age_v_price(df, var)
```

Age of the house and the age since some home improvement seems to be a significant factor that effects sale price.

### A look at  Numerical categorical variables

```python
cat_vars = [var for var in num_vars if len(df[var].unique())<20 and var not in YrVar + ['Id']]
```

```python
df[cat_vars].head()
```

```python
def cat_v_price(data, var):
    """
    Compares categories within a variable to the sale price.

    inputs:
    data: A dataframe
    var: A variable name from the same dataframe
    outputs:
    a scatter plot
    """

    data = data.copy()
    
    data.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.ylabel('SalePrice')
    plt.show()
    
for var in cat_vars:
    cat_v_price(df, var)
```

There are variables where the category makes a lot of difference like the variable OverallQual. Whoever was the decider of overall quality was apparently very good at it because the result was a very clean relationship between this measure and the sale price. For things such as 'BsmtHalfBath' there is little correlation between this variable and the sale price.

### A look at numerical continous Variables

```python
cont_vars = [var for var in num_vars if var not in cat_vars+YrVar+['Id']]
df[cont_vars].head()
```

```python
def cont_dist(data, var):
    """
    Checks data distribution of continous variables.

    inputs:
    data: A dataframe
    var: A variable name from the same dataframe
    outputs:
    a distribution plot
    """

    data = data.copy()
    
    data[var].hist(bins=25)
    plt.title(var)
    plt.xlabel(var)
    plt.ylabel('# of houses')
    plt.show()
    
for var in cont_vars:
    cont_dist(df, var)
```

As we can see most of these variables are right skewed which isn't too surprising because a house can only be so modest and most houses will be more modest than gaudy. Since I plan to use a linear model to model this I should normalize this data.

#### Log Normalization of continous numerical data

```python
def cont_dist_transformed(data, var):
    """
    Checks data distribution of continous variables.

    inputs:
    data: A dataframe
    var: A variable name from the same dataframe
    outputs:
    a distribution plot
    """

    data = data.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        data[var] = np.log(data[var])
        data[var].hist(bins=25)
        plt.title(var)
        plt.xlabel(var)
        plt.ylabel('# of houses')
        plt.show()
    
for var in cont_vars:
    cont_dist_transformed(df, var)
```

In the data that I could transform there the log transformation fixed the right skew. Now lets see the relationship between these transformed data points and the sale price

```python
def cont_transformed_v_price(data, var):
    """
    Checks continous variables to the price data.

    inputs:
    data: A dataframe
    var: A variable name from the same dataframe
    outputs:
    a scatter plot
    """

    data = data.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        data[var] = np.log(data[var])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[var], data['SalePrice'])
        plt.xlabel(var)
        plt.ylabel('Sale Price')
        plt.show()
    
for var in cont_vars:
    if var != 'SalePrice':
        cont_transformed_v_price(df, var)
```

The first two plots don't show a clear relationship but there is an obvious linear relationship to exploit for the variables '1stFirSF' and 'GrLivArea'

### Outliers

```python
def find_outliers(data, var):
    """
    Looks for outliers in data with non zero values.

    inputs:
    data: A dataframe
    var: A variable name from the same dataframe
    outputs:
    a boxplot
    """

    data = data.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        data[var] = np.log(data[var])
        data.boxplot(column=var)
        plt.title(var)
        plt.ylabel(var)
        plt.show()
    
for var in cont_vars:
        find_outliers(df, var)
```

There are a few outliers in each variable with considerabley more in the lot area than the others. They are still few enough that I will ignore them for the first round of modeling.

### Rare Labels

```python
def analyze_rare_labels(data, var, rare_pct):
    """
    Designed to find exceptional rare labels within a categorical variable
    
    inputs:
    data: A dataframe(df)
    var: A variable name from the same dataframe(str)
    rare_pct: A percentage that is a boundry for whether a categorical
    variables occurence is rare or not(float)
    outputs:
    a string showing which label within each category appears below the rare_pct threshold
    """
    data = data.copy()
    tmp = data.groupby(var)['SalePrice'].count()/len(data)
    return tmp[tmp<rare_pct]

for var in cat_var:
    print(analyze_rare_labels(df, var, 0.01))
    print()
```

There are plenty of categorical values that appear so rarely within the dataset that it will be borderline impossible to derive any real insight from them. This will be dealt with in the feature engineering section.
